```python
# count plot:
countplot plots the count of the number of records by category
# Bar plot
Barplot plots a value or metric for each category(by default, barplot plots the mean of a variable,bycategory )
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.countplot(x="sex",data=var)
plt.show() # countplot gives the count on the basis of filter condition
```

```python
sns.barplot(x="sex",y="size",data=var)
plt.show() # where as barplot gives the mean of the records or data
```

```python
sns.countplot(x="sex",data=var,hue="smoker")
plt.show() # countplot gives the count on the basis of filter condition
```

```python
sns.countplot(y="sex",data=var,hue="smoker")
plt.show() # countplot gives the count on the basis of filter condition
```

```python
sns.countplot(x="sex",data=var,hue="smoker",palette ="bwr" )
plt.show() # countplot gives the count on the basis of filter condition
```

```python
sns.countplot(x="sex",data=var,hue="smoker",color = "y")
plt.show() # countplot gives the count on the basis of filter condition
```

```python
sns.countplot(x="sex",data=var,hue="smoker",color = "y",saturation = 0.3)
plt.show() # countplot gives the count on the basis of filter condition
```

# Violin plot in seaborn

A violin plot plays a similar role as a box and whisker plot. it shows the distribution of quantitative data across several levels of one(or more) categorical variables such that those distributions can be compared

import seaborn 

seaborn.set(style = 'whitegrid')

f = seaborn.load_dataset("fmri")

seaborn.violinplot(x="",y="",data = f)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

wall = sns.load_dataset("tips").head(20)
wall
```

```python
sns.pairplot(wall)
plt.show()
```

```python
sns.pairplot(wall,hue = "sex")
plt.show()
```

```python
sns.pairplot(wall,vars=["total_bill","tip"],hue = "sex")
plt.show()
```

```python
sns.pairplot(wall,vars=["total_bill","tip"],hue = "sex",hue_order = ["Female","Male"])
plt.show()
```

```python
sns.pairplot(wall,vars=["total_bill","tip"],hue = "sex",hue_order = ["Female","Male"],palette = "BuGn")
plt.show()
```

```python
sns.pairplot(wall,hue = "sex",hue_order = ["Female","Male"],x_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(wall,hue = "sex",hue_order = ["Female","Male"],x_vars=["total_bill","tip"],kind  ="reg")
plt.show()
```

```python
sns.pairplot(wall,hue = "sex",hue_order = ["Female","Male"],kind  ="reg") # kind are scatter,   reg,  kde, hist
plt.show()
```

```python
sns.pairplot(wall,hue = "sex",hue_order = ["Female","Male"],kind  ="kde")
plt.show()
```

```python
sns.pairplot(wall,hue = "sex",hue_order = ["Female","Male"],kind  ="kde",diag_kind ="hist")
plt.show()
```

```python
sns.pairplot(wall,hue = "sex",hue_order = ["Female","Male"])
plt.show()
```

```python
sns.pairplot(wall,hue = "sex",hue_order = ["Female","Male"],markers=["*",">"])
plt.show()
```

# Strip plot in seaborn

A Strip plot is very simple to understand. it basically a scatter plot that differentiates different categories.
strip plots are considered a good alternative to a boxplot or a violin plot

import seaborn as sns

import matplotlib.pyplot as plt

tip = sns.load_dataset("")
sns.striplot(x="",y="",data=tip)

plt.show()

```python
import seaborn as sns
import matplotlib.pyplot as plt

vel = sns.load_dataset("tips")
vel
```

```python
sns.stripplot(x="day",y="total_bill",data = vel)
```

```python
sns.stripplot(x="day",y="total_bill",data = vel,hue="sex")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data = vel,hue="sex",palette = "rocket_r")
plt.show()
```

# Box plot in seaborn

# Syntax
import seaborn as sns
import matplotlib.pyplot as plt

f = sns.load_dataset("")

sns.boxplot(x="",y="",data=f)
plt.show()

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

f = sns.load_dataset("tips")
f
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=f)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=f,hue = "time")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=f,hue = "sex")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=f,hue = "sex",color="g")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=f,hue = "sex",color="g",order=["Fri","Sun","Thur","Sat"])
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=f,hue = "sex",color="g",order=["Fri","Sun","Thur","Sat"],showmeans = True)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=f,color="r",order=["Fri","Sun","Thur","Sat"],showmeans = True)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=f,color="w",order=["Fri","Sun","Thur","Sat"],showmeans = True,
            meanprops={"marker":">","markeredgecolor":"b"},linewidth=3)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=f,order=["Fri","Sun","Thur","Sat"],showmeans = True,
            meanprops={"marker":">","markeredgecolor":"b"},linewidth=3)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=f,order=["Fri","Sun","Thur","Sat"],showmeans = True,
            meanprops={"marker":">","markeredgecolor":"b"},linewidth=3,palette="coolwarm")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=f,order=["Fri","Sun","Thur","Sat"],showmeans = True,
            meanprops={"marker":">","markeredgecolor":"b"},linewidth=3,palette="coolwarm",orient = "v")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data = vel,showmeans = True,
            meanprops={"marker":"*","markeredgecolor":"b"},linewidth=3,palette="coolwarm",orient = "h")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x=vel["total_bill"])

plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(y=vel["total_bill"])

plt.show()
```
