```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset("")

sns.lineplot(x="",y="",data=data)

plt.show()
```

```python
y_1 = sns.load_dataset("penguins")
y_1
```
