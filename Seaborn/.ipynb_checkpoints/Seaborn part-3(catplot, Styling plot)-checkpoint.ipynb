import seaborn as sns

ex = sns.load_dataset("")

h = sns.boxplot(x="",y="",data=ex)

```python
import seaborn as sns
import matplotlib.pyplot as plt

ex = sns.load_dataset("tips").head(30)
ex
```

```python
sns.catplot(x = "tip",y = "size",data = ex)
plt.show()
```

```python
sns.catplot(x = "tip",y = "size",data = ex,hue="sex")
plt.show()
```

```python
sns.catplot(x = "tip",y = "size",data = ex,hue="sex",palette = "gnuplot")
plt.show()
```

```python
sns.catplot(x = "tip",y = "size",data = ex,hue="sex",palette = "gnuplot",height=3)
plt.show()
```

```python
sns.catplot(x = "tip",y = "size",data = ex,hue="sex",palette = "Oranges",kind="bar")
plt.show()
```

```python
sns.catplot(x = "day",y = "size",data = ex,hue="sex",palette = "Oranges",kind="point")
plt.show()
```

```python
sns.catplot(x = "day",y = "size",data = ex,hue="sex",palette = "Oranges",kind="box")
plt.show()
```

```python
sns.catplot(x = "day",y = "size",data = ex,hue="sex",palette = "Oranges",kind="boxen")
plt.show()
```

# Styling plot in seaborn

```python
# To configure seaborn figure styles
# To Remove axes spines
# To scale and context
```

```python
ex1 = sns.load_dataset("tips")
ex1
```

```python
sns.boxplot(y=ex1["total_bill"])
```

```python
sns.barplot(x = "day",y="total_bill",data = ex1)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.barplot(x = "day",y="total_bill",data = ex1)
plt.grid()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(x = "day",y="total_bill",data = ex1)
sns.despine()
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(3,8))
sns.barplot(x = "day",y="total_bill",data = ex1)
sns.despine()
plt.show()
```

```python
sns.set_style("whitegrid")

sns.set_context("poster",font_scale=0.5)
sns.barplot(x = "day",y="total_bill",data = ex1)

plt.show()
```

```python
sns.set_style("whitegrid")

sns.set_context("paper",font_scale=0.5)
sns.barplot(x = "day",y="total_bill",data = ex1)

plt.show()
```

```python
sns.set_style("whitegrid")

sns.set_context("paper",font_scale=0.5)
sns.barplot(x = "day",y="total_bill",data = ex1,palette = "cool")

plt.show()
```

# multi plot(Facet - Grid) in seaborn

# Syntax
import seaborn as sns

import matplotlib.pyplot as plt

df = seaborn.load_dataset('')

graph = seaborn.facetgrid(df,col=" ",hue=" ")

graph.map(plt.scatter).add_legend()

plt.show()


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ex1
```

```python
fg = sns.FacetGrid(ex1)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg = sns.FacetGrid(ex1,col = "sex")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg = sns.FacetGrid(ex1,col = "sex",hue="day")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg = sns.FacetGrid(ex1,col = "sex",hue="day")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(ex1,col = "day",hue="sex")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(ex1,col = "day",hue="sex")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(ex1,col = "sex",hue="day")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(ex1,col = "sex",hue="day",palette = "summer")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(ex1,col = "sex",hue="day",palette = "summer")
fg.map(plt.bar,"total_bill","tip",edgecolor ="r").add_legend()
plt.show()
```
