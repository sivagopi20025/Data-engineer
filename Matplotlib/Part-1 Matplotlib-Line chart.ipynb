```python
# Normal Bar plot
A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]

plt.bar(A,B)
plt.show()
```

```python
# Bar plot with Distingushable colors on each bar
A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]
c = ['g','b','y','r']

plt.bar(A,B,color = c)
plt.show()
```

```python
# Create labels on x,y-axis of the plot along with color
A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]
c = ['g','b','y','r']
plt.xlabel("Language")
plt.ylabel("No. ")


plt.bar(A,B,color = c)
plt.show()
```

```python
# Create labels on title of the plot 
A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]
c = ['g','b','y','r']
plt.xlabel("Language")
plt.ylabel("No. ")
plt.title("Language popularity")
plt.bar(A,B,color = c)
plt.show()
```

```python

#  Change Font size of all labels
A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]
c = ['g','b','y','r']
plt.xlabel("Language",fontsize = 20)
plt.ylabel("No. ",fontsize = 20)
plt.title("Language popularity",fontsize = 20)
plt.bar(A,B,color = c)
plt.show()
```

```python

#  Change width of graph bar in plot
A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]
c = ['g','b','y','r']
plt.xlabel("Language",fontsize = 20)
plt.ylabel("No. ",fontsize = 20)
plt.title("Language popularity",fontsize = 20)
plt.bar(A,B,width = 0.5,color = c)
plt.show()
```

```python

#  Change bar plots bar position here it is set on edge with align
A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]
c = ['g','b','y','r']
plt.xlabel("Language",fontsize = 20)
plt.ylabel("No. ",fontsize = 20)
plt.title("Language popularity",fontsize = 20)
plt.bar(A,B,width = 0.5,color = c,align = 'edge')
plt.show()
```

```python

#  Change bar plots bar position here it is set on center with align
A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]
c = ['g','b','y','r']
plt.xlabel("Language",fontsize = 20)
plt.ylabel("No. ",fontsize = 20)
plt.title("Language popularity",fontsize = 20)
plt.bar(A,B,width = 0.5,color = c,align = 'center')
plt.show()

# by default align position of bar is center
```

```python
# multiple variables with different popularity data in bar plot


A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]
D = [20,30,40,50]
# c = ['g','b','y','r']
plt.xlabel("Language",fontsize = 20)
plt.ylabel("No. ",fontsize = 20)
plt.title("Language popularity",fontsize = 20)
plt.bar(A,B,width = 0.5,color = "r",edgecolor = 'k',label = "Programm language") 
plt.bar(A,D,width = 0.5,color = "y",edgecolor = 'r',label = "Programm language1") 
plt.legend()
plt.show()
```

```python
# multiple variables with different popularity data in bar plot
import matplotlib.pyplot as plt
import numpy as np
A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]
D = [20,30,40,50]

width = 0.3
p = np.arange(len(A))
p1 = [j+width for j in p]

plt.xlabel("Language",fontsize = 20)
plt.ylabel("No. ",fontsize = 20)
plt.title("Language popularity",fontsize = 20)

plt.bar(p,B,width,color = "r",edgecolor = 'k',label = "Programm language") 
plt.bar(p1,D,width,color = "y",edgecolor = 'r',label = "Programm language1") 

plt.xticks(p+width/2,A,rotation = 45)

plt.legend()
plt.show()
```

```python
# multiple variables with different popularity data with changing bar width in bar plot


A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]
D = [20,30,40,50]
# c = ['g','b','y','r']
width = 0.3
plt.xlabel("Language",fontsize = 20)
plt.ylabel("No. ",fontsize = 20)
plt.title("Language popularity",fontsize = 20)
plt.barh(A,B,width,color = "r",edgecolor = 'k',label = "Programm language") 
plt.barh(A,D,width,color = "y",edgecolor = 'r',label = "Programm language1") 
plt.legend()
plt.show()
```

```python
# multiple variables with different popularity data in bar plot
import matplotlib.pyplot as plt
import numpy as np
A = ["python","c-language","C++","Scala"]
B = [85,70,60,82]
D = [20,30,40,50]

width = 0.2

plt.xlabel("Language",fontsize = 20)
plt.ylabel("No. ",fontsize = 20)
plt.title("Language popularity",fontsize = 20)

plt.barh(A,B,width,color = "r",label = "Programm language") 
plt.barh(A,D,width,color = "y",label = "Programm language1") 

plt.legend()
plt.show()
```
