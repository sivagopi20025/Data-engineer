# Fill_between

```python
d = [1,2,3,4,5]
e = [1,2,3,4,5]


plt.plot(d,e,color="r")

plt.fill_between(x=[2,4],y1=2,y2=4)

plt.title("python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

d=np.array([1,2,3,4,5])
e=np.array([1,2,3,4,5])

plt.plot(d,e,color="r")

plt.fill_between(d,e,color="g",where=(d >= 2) & (d <= 4))

plt.title("python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

d=np.array([1,2,3,4,5])
e=np.array([1,2,3,4,5])

plt.plot(d,e,color="r")

plt.fill_between(d,e,color="g",where=(d >= 2) & (d <= 4),alpha=0.5)

plt.title("python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.show()
```
