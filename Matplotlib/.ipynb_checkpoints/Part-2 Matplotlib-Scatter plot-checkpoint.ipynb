# Scatter plot Syntax

import matplotlib.pyplot as plt

x = []
y = []

plt.show()

```python
# Basic Scatter plot

import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]

plt.scatter(x,y)
plt.show()
```

```python
# Basic Scatter plot with label outside the perimeter(title)


x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]

plt.scatter(x,y)
plt.title("scatter Plot", fontsize = 15)
plt.show()
```

```python
# Basic Scatter plot with x,y labels as day and no(number) with title

Day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]

plt.scatter(Day,no)
plt.title("scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

```python
# Basic Scatter plot with x,y labels as day and no(number) with title and color change of the values inside plot

Day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]

plt.scatter(Day,no,color = "r")
plt.title("scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

```python
# Basic Scatter plot with x,y labels as day and no(number) with title

Day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = ["r","y","g","b","r","g","r"]
plt.scatter(Day,no,c = colors)
plt.title("scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

```python
# Basic Scatter plot with x,y labels as day and no(number) with title and size of dots

Day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = ["r","y","g","b","r","g","r"]
sizes = [200,300,100,400,500,300,600]
plt.scatter(Day,no,c = colors,s = sizes)
plt.title("scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

```python
# Basic Scatter plot with x,y labels as day and no(number) with title and size of dots

Day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = ["r","y","g","b","r","g","r"]
sizes = [200,300,100,400,500,300,600]
plt.scatter(Day,no,c = colors,s = sizes,alpha = 0.4)
plt.title("scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

```python
# Scatter plot with x,y labels with colors and transparent dots

Day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = ["r","y","g","b","r","g","r"]
sizes = [200,300,100,400,500,300,600]
plt.scatter(Day,no,c = colors,s = sizes,alpha = 0.4,marker = "*")
plt.title("scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

```python
# Scatter plot with x,y labels with colors and transparent dots

Day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = ["r","y","g","b","r","g","r"]
sizes = [200,300,100,400,500,300,600]
plt.scatter(Day,no,c = colors,s = sizes,edgecolor = "g",linewidth = 2)
plt.title("scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

```python
# Scatter plot with x,y labels with colors and transparent dots

Day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = [20,30,40,45,50,60,75]
sizes = [200,300,100,400,500,300,600]
plt.scatter(Day,no,c = colors,s = sizes,cmap = "BrBG",alpha = 0.5)
t = plt.colorbar()
t.set_label("ColorBar")
plt.title("scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

```python
# Scatter plot with x,y labels with colors and transparent dots

Day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = [20,30,40,45,50,60,75]
sizes = [200,300,100,400,500,300,600]
plt.scatter(Day,no,c = colors,s = sizes,cmap = "viridis",alpha = 0.5)
t = plt.colorbar()
t.set_label("ColorBar",fontsize = 15)
plt.title("scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

```python
# Scatter plot with x,y labels with colors and transparent dots

Day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
noo = [3,2,4,5,1,6,2]
colors = [20,30,40,45,50,60,75]
sizes = [200,300,100,400,500,300,600]
plt.scatter(Day,no,c = colors,s = sizes,cmap = "viridis",alpha = 0.5)
plt.scatter(Day,noo,c = "r",s = sizes,cmap = "viridis")
t = plt.colorbar()
t.set_label("ColorBar",fontsize = 15)
plt.title("scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```
