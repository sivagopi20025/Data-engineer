```python
import matplotlib.pyplot as plt
# How to increase or decrease the radius of the pie chart
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ['c','c++','Java','Python']

c = ["g","y","orange","b"]

plt.pie(x,labels=y,radius = 1.5)
cr = plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)

plt.show()

```
