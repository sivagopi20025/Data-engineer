```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")

plt.subplot(2,2,2)
plt.pie([1],colors="r")

x1 = [10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x)

plt.subplot(2,2,4)
x2 = ["a","s","d","f"]
y2 = [10,20,30,40]
plt.bar(x2,y2)

plt.show()
```

# savefig

# Syntax

import matplotlib.pyplot as plt

x = []

y = []

plt.plot(x,y)

plt.savefig(fname)

plt.show()

```python
import matplotlib.pyplot as plt


x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("basic line chart",dpi = 2000)
plt.show()
```

```python
import matplotlib.pyplot as plt


x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("basic line chart",dpi=2000,facecolor='y')
plt.show()
```

```python
import matplotlib.pyplot as plt


x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("basic line chart.pdf",dpi=2000,facecolor='y')
plt.show()
```

```python
import matplotlib.pyplot as plt


x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("basic linechart.pdf",dpi=2000,facecolor='y',transparent=True)
plt.show()
```

```python
import matplotlib.pyplot as plt


x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("linechart.pdf",dpi=2000,facecolor='y',transparent=True,bbox_inches="tight")
plt.show()
```

# Axis

# xticks(),yticks(),xlim(),ylim(),axis()

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x,labels=["python","c","c++","java","scala"])
plt.yticks(x)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)
plt.xlim(0,10)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.xlim(0,10)
plt.ylim(0,10)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.axis([0,10,0,7])

plt.show()
```

# Working with text in matplotlib python

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title("everything",fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title("everything",fontsize=20)
plt.xlabel("Days")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title("everything",fontsize=20)
plt.xlabel("Days",fontsize=20)
plt.ylabel("Python",fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title("everything",fontsize=20)
plt.xlabel("Days",fontsize=20)
plt.ylabel("Python",fontsize=20)

plt.text(2,5,"pytorch",fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title("everything",fontsize=20)
plt.xlabel("Days",fontsize=20)
plt.ylabel("Python",fontsize=20)

plt.text(2,5,"pytorch",fontsize=15,style="italic",bbox = {"facecolor":"red"})

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title("everything",fontsize=20)
plt.xlabel("Days",fontsize=20)
plt.ylabel("Python",fontsize=20)

plt.text(2,5,"pytorch",fontsize=15,style="italic",bbox = {"facecolor":"red"})
plt.annotate("python",xy=(2,1),xytext = (3,3),arrowprops = dict(facecolor= "black",shrink=100))

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title("everything",fontsize=20)
plt.xlabel("Days",fontsize=20)
plt.ylabel("Python",fontsize=20)

# plt.text(2,5,"pytorch",fontsize=15,style="italic",bbox = {"facecolor":"red"})
# plt.annotate("python",xy=(2,1),xytext = (3,3),arrowprops = dict(facecolor= "black",shrink=100))

plt.legend(["up"],loc=0,facecolor="red",edgecolor="black",framealpha=0.2,shadow=True)
plt.show()
```
