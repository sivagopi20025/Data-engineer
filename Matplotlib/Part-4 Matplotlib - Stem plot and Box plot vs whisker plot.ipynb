```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5,6]
b = [2,2,5,6,4,3]

plt.stem(a,b)
plt.show()
```

# for changing the solid line to dotted lin stem plot use Linefmt

Linefmt                 character                linestyle

                         "'_'"                  Solid line
                         
                         "'__`"                 Dashed line
                         
                         "'_.'"                 Dash-dot line
                         
                         "':'"                  Dotted line




```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5,6]
b = [2,2,5,6,4,3]

plt.stem(a,b)
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5,6]
b = [2,2,5,6,4,3]

plt.stem(a,b,linefmt=":")
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5,6]
b = [2,2,5,6,4,3]

plt.stem(a,b,linefmt=":",markerfmt="ro")
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5,6]
b = [2,2,5,6,4,3]

plt.stem(a,b,linefmt=":",markerfmt="ro",bottom = 0)
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5,6]
b = [2,2,5,6,4,3]

plt.stem(a,b,linefmt=":",markerfmt="ro",bottom = 0,basefmt="g")
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5,6]
b = [2,2,5,6,4,3]

plt.stem(a,b,linefmt=":",markerfmt="ro",bottom = 0,basefmt="g",label="python")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5,6]
b = [2,2,5,6,4,3]

plt.stem(a,b,linefmt=":",markerfmt="ro",bottom = 0,basefmt="g",label="python",use_line_collection=False)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5,6]
b = [2,2,5,6,4,3]

plt.stem(a,b,linefmt=":",markerfmt="ro",bottom = 0,basefmt="g",label="python",use_line_collection=False,orientation="horizontal")
plt.legend()
plt.show()
```

# Box Plot

```python
# Basic Box Plot
import matplotlib.pyplot as plt

x = [2,4,6,10,14,18,20,26,30,60]

plt.boxplot(x)

plt.show()
```

```python

x = [10,20,30,40,50,60,70]

plt.boxplot(x)

plt.show()
```

```python

x = [10,20,30,40,50,60,70]
# To show the box pot with different shape other than box use notch as True
plt.boxplot(x,notch=True)

plt.show()
```

```python

x = [10,20,30,40,50,60,70]
# to show plot in horizontally
plt.boxplot(x,vert=False)

plt.show()
```

```python

x = [10,20,30,40,50,60,70]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,vert=False,widths = 0.4) # default width = 0.2

plt.show()
```

```python

x = [10,20,30,40,50,60,70]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,vert=False,widths = 0.4,labels=["Python"],patch_artist = True ) # default width = 0.2

plt.show()
```

```python

x = [10,20,30,40,50,60,70]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,vert=True,widths = 0.2,showmeans=True) # default width = 0.2

plt.show()
```

```python

x = [10,20,30,40,50,60,70]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,vert=False,widths = 0.4,labels=["Python"],patch_artist = False,showmeans=True ) # default width = 0.2

plt.show()
```

```python

x = [10,20,30,40,50,60,70,120]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,vert=False,widths = 0.4,labels=["Python"],patch_artist = False,showmeans=True ) # default width = 0.2

plt.show()
```

```python

x = [10,20,30,40,50,60,70,120]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,vert=False,widths = 0.4,labels=["Python"],patch_artist = False,showmeans=True,whis=1) # default width = 0.2

plt.show()
```

```python

x = [10,20,30,40,50,60,70,120]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,vert=False,widths = 0.4,labels=["Python"],patch_artist = False,showmeans=True,whis=1,sym="g*") # default width = 0.2

plt.show()
```

```python

x = [10,20,30,40,50,60,70,120]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,vert=False,widths = 0.4,labels=["Python"],patch_artist = False,showmeans=True,whis=1,sym=" ") # default width = 0.2

plt.show()
```

```python

x = [10,20,30,40,50,60,70,120]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,vert=False,widths = 0.4,labels=["Python"],patch_artist = False,showmeans=True,whis=1,sym="g*",
            boxprops=dict(color="r")) # default width = 0.2

plt.show()
```

```python

x = [10,20,30,40,50,60,70,120]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,vert=False,widths = 0.4,labels=["Python"],patch_artist = False,showmeans=True,whis=1,sym="g*",
            boxprops=dict(color="r"),capprops=dict(color="r"))

plt.show()
```

```python

x = [10,20,30,40,50,60,70,120]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,vert=False,widths = 0.4,labels=["Python"],patch_artist = False,showmeans=True,whis=1,sym="g*",
            boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="r"))

plt.show()
```

```python

x = [10,20,30,40,50,60,70,120]
# to show plot in horizontally and increase the width of box
plt.boxplot(x,
            vert=False,
            labels=["Python"],
            showmeans=True,whis=1,
            sym="g*",
            boxprops=dict(color="r"),
            capprops=dict(color="r"),
            whiskerprops=dict(color="r"),
            flierprops=dict(markeredgecolor="b"))

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(
    x,
    vert=False,
    labels=["Python"],
    showmeans=True,
    whis=1,
    sym="g*",
    boxprops=dict(color="r", linewidth=2),    # Added linewidth for wider boxes
    capprops=dict(color="r"),
    whiskerprops=dict(color="r"),
    flierprops=dict(markeredgecolor="b")
)

plt.show()
```
